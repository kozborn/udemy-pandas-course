## .query() method (doesn't work when there is a space in column name)
so we need to replace it

```python
bond.columns = [column.replace(' ', '_') for column in bond.columns]
bond.head()

bond.query('Actor == "Sean Connery"')
office300 = bond.query('Box_Office > 300').sort_values(by="Box_Office")
office300.head()
```

## .apply() method

```python
bond = pd.read_csv('./data/jamesbond.csv', index_col="Film")
bond.sort_index(inplace=True)

def addMilions(number):
    return str(number) + " Milions"

columnsToChange = ['Box Office', 'Budget']

# bond['Box Office'] = bond['Box Office'].apply(addMilions)
# bond['Budget'] = bond['Budget'].apply(addMilions)

for col in columnsToChange:
    bond[col] = bond[col].apply(addMilions)
bond.head()
```

## .apply() method on rows


```python
bond = pd.read_csv('./data/jamesbond.csv', index_col="Film")
bond.sort_index(inplace=True)
bond.head()
```

```python
def goodMovie(row):
    actor = row[1]
    budget = row[4]
    if actor == 'Pierce Brosnan':
        return 'The best'
    if actor == 'Roger Moore' and budget > 50:
        return 'Enjoyable'
    return 'I have no clue'


bond.apply(goodMovie, axis=1)
    
```
